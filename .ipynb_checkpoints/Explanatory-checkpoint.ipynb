# Insights out of 5 millions rides by GoFordBike

## Investigation Overview

In this investigation, I wanted to visualize the the relationship between duration of the trip and the covered distance for GoFordBike, and to know when did GoFordBike got many rides and when it didn't, also to know the affect of user type on duration and distance of trip.

## Dataset Overview

The dataset contains 26 separate files downloaded from GoFordBike website, each file contains all rides data in a specific month from July 2017 to January 2020, the rides data contains of ride duration, starting and ending locations and timing and the user type. After combining all the 26 files, the combined data counted more than 5 millions rows.

## Data Wrangling

```python
# import all packages and set plots to be embedded inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import os
import geopy.distance

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in all the datasets for all months, and store all datasets in a list of dataframes
data_dfs = []
for file in os.listdir("./data files"):
    if file.endswith(".csv"):
        print(file)
        file = os.path.join('.\data files',file)
        df = pd.read_csv(file)
        data_dfs.append(df)
```

```python
#Drop the 2 cols that don't present in some dfs, as they are not very important to the analysis
for df in data_dfs:
    if 'bike_share_for_all_trip' in list(df):
        df.drop(columns=['bike_share_for_all_trip'], inplace=True)
    if 'rental_access_method' in list(df):
        df.drop(columns=['rental_access_method'], inplace=True)
```

```python
#Add all dfs in only one df
i = 1
for df in data_dfs:
    print('Dataframe ', i, ' is being added ...' )
    data_dfs[0] = data_dfs[0].append(df, ignore_index=True)
    print('Done')
    i = i + 1
```

```python
#Store the combined df in a variable outside the list
df = data_dfs[0]
```

```python
df_improved = df[df.duration_sec < 90000]
```

## Distribution of Rides Duration

Rides durations in the dataset take on a very large range of values, from 60 seconds as minimum which is propably due to standard values for shortest rides in the system, and ranges to about 86k seconds at the highest. Plotted on a logarithmic scale, the distribution of durations is normally distributed with small right skew.

```python
log_binsize = 0.1
bins = 10 ** np.arange(np.log10(60), np.log10(df_improved.duration_sec.max())+log_binsize, log_binsize)

plt.figure(figsize=[8, 5]);
plt.hist(data = df_improved, x = 'duration_sec', bins = bins);
plt.xscale('log');
plt.xlabel('Duration in m (minutes) or h (hours)');
ticks = [60, 120, 180, 480, 960, 1980, 5400, 9900, 19800, 49680, 99720]
ticks_labels = ['1m', '2m', '3m', '8m', '16m', '33m', '1.5h', '2.75h', '5.5h', '13.8h', '27.7h']
ticksy = [0, 100000, 200000, 300000, 400000, 500000, 600000, 700000, 800000]
ticks_labelsy = ['0', '100k', '200k', '300k', '400k', '500k', '600k', '700k', '800k']
plt.xticks(ticks, ticks_labels, rotation=15);
plt.yticks(ticksy, ticks_labelsy);
med = plt.axvline(x=df_improved.duration_sec.median(), c='red');
mean = plt.axvline(x=df_improved.duration_sec.mean(), c='black');
plt.title('Distribution of Rides Durations')
plt.legend((med, mean), ('Median of Duration (9.5 mins)', 'Mean of Duration (14.6 mins)'))
plt.show();
```

## Distribution of rides displacement in a 1000 rides sample 

Displacment is measured between the starting and ending coordinates points, and then a sample of 10,000 rides were chosen randomly to plot the distribution of the displacement values in km.

```python
def get_distance(ride):
    coords_1 = [ride.start_station_latitude, ride.start_station_longitude]
    coords_2 = [ride.end_station_latitude, ride.end_station_longitude]
    return geopy.distance.geodesic(coords_1, coords_2).km
```

```python
samples = np.random.choice(df_improved.shape[0], 1000, replace = False);
df_samp = df_improved.loc[samples,:];
```

```python
df_samp['distance_km'] = df_samp.apply(get_distance, axis=1);
```

```python
bins = np.arange(0,df_samp.distance_km.max()+0.1, 0.1)

plt.hist(data=df_samp, x='distance_km', bins=bins, weights = np.zeros_like(df_samp.distance_km) + 1. / df_samp.shape[0]);

plt.xlabel('Displacement (KM)');

plt.ylabel('Relative Frequency')
plt.xlim(0,8)
plt.title('Distribution of rides displacements in 10000 rides sample')
plt.show();
```

-----------------------------------

## Relation between duration and displacements in a 1000 rides sample 

The relations has a strong positive correlation, however many points residing at zero displacements are having large duration values, and this is because the fact that some riders pick up the bikes from a certain station and after the trip finish, the rider drops off at the same station.

```python
sb.regplot(data = df_samp, x = 'distance_km', y = 'duration_sec');
plt.ylim(0,20000);
plt.xlabel('Displacement Km')
plt.ylabel('Duration Hours')
ticks = [0, 2500, 5000, 7500, 10000, 12500, 15000, 17500, 20000]
ticks_labels = ['0', '0.70', '1.39', '2.08', '2.78', '3.47', '4.17', '4.86', '5.56']
plt.yticks(ticks, ticks_labels);
plt.title('Relation between duration and displacement');
```

## Number of rides per month between July 2017 and January 2020 

The average number of rides per month is 187k rides, and the year 2019 is showing great results in against 2018 in terms of number of rides per month.

Also, Decemeber is always associated with lower number of rides per month, which can show that rides decrease in holidays, which also can lead to a conclusion that people use these rides during working times.

```python
# Change the type of time to be datetime type
df_improved.start_time = pd.to_datetime(df_improved.start_time)
df_improved.end_time = pd.to_datetime(df_improved.end_time)

df_time_indexed = df_improved.copy()

df_time_indexed.index = df_time_indexed['start_time']
rides_per_month = df_time_indexed.resample('M').duration_sec.count()

rides_per_month = rides_per_month[1:]
```

```python
# Plot a time line graph that shows the number of rides per month
plt.figure(figsize=(10,5));
rides_per_month.plot();
plt.xlabel('Month')
plt.title('Number of rides per month')
plt.yticks(np.arange(0, 35e4, 5e4), ['0', '50k', '100k','150k','200k','250k','300k']);
axmean = plt.axhline(y=rides_per_month[1:].mean(), c='red');
plt.legend([axmean], ['Mean of number of rides per month (187k rides)']);
plt.grid(which='both')
```
